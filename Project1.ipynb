```python
import pandas as pd
```

```python
path = ("athlete_events.csv")
df_events = pd.read_csv(path)
df_events["join_id"] = df_events["NOC"] + df_events["Year"].astype(str)
df_events.head(10)

```

```python
path2 = ("population_by_country.csv")
df_pop = pd.read_csv(path2)
df_pop["join_id"] = df_pop["NOC"] + df_pop["Year"].astype(str)
df_pop.head(10)
```

```python
merge_df = pd.merge(df_events, df_pop, on="join_id")
merge_df.shape
```

```python
# Export file as a CSV, without the Pandas index, but with the header
merge_df.to_csv("output/merged_data.csv", index=False, header=True)
```
